<h4>Efficient Markets, Random Walk, Chaos Theory,
and “Nature’s Law”</h4>

* Efficient Market theory is an academic notion that nobody can outperform the market because any price at any given moment incorporates all available information.
* Random Walk theorists claim that market prices change at random. 
* Chaos Theory has achieved prominence in the recent decades. Markets are largely chaotic, and the only time you can have an edge is during orderly periods.
* Nature’s Law is the rallying cry of a clutch of mystics who claim there is a perfect order in the markets 

<h3>18. Support and Resistance</h3>

* __Support__ (__Resistance__) is a price level where buying (selling) is strong enough to interrupt or reverse a downtrend (uptrend). Support (resistance) is represented on a chart by a horizontal line connecting __two or more bottoms__ .
*  Support and resistance areas often __switch roles__.
* The strength of these barriers __increases__ each time prices touch them and bounce away.
* Beware of __false breakouts__ from support and resistance. Amateurs tend to __follow breakouts__, while professionals tend to __fade (trade against) them__.

<h4>True and False Breakouts</h4>

* Markets spend more time in trading ranges than in trends. Most breakouts from trading ranges are false breakouts.
* Professionals expect prices to fluctuate most of the time, without going anywhere far.
* The best time to buy an upside breakout on a daily chart is when your analysis of the weekly chart suggests that a new uptrend is developing.
*  True breakouts are confirmed by heavy volume, while false breakouts tend to have light volume

<h3>19. Trends and Trading Ranges<h5>p.60</h5></h3>

* In a trading range, most rallies stop at about the same high level, and declines peter out at about the same low level. 

* __Trends__ and __trading ranges__ call for different tactics. 
<ol>
    <li>When you go long in an uptrend or sell short in a downtrend, you have to give that trend the benefit of
the doubt and use a wider stop, so as not to be shaken out easily. You have to <b>follow strength during trends</b>—buy in uptrends and short in downtrends.</li>
    <li>In a trading range, on the other hand, you have to use tight stops, be nimble and close out positions at
        the slightest sign of a reversal. You aim to <b>buy weakness</b> and <b>sell strength</b></li>

</ol>

<h4>Trade or Wait</h4>

Having identified an uptrend, you need to decide whether 
* to buy immediately or
    * If you buy fast, you’ll get in gear with the trend, but on the minus side, your stops are likely to be farther away, increasing your risk.
* wait for a dip
    * If you wait for a dip, your risk will be smaller, but you’ll have four groups of competitors: 
        <ol>
            <li>longs who want to add to their positions,</li>
            <li>shorts who want to get out even,</li>
            <li>traders who never bought (such as yourself),</li>
            <li>and traders who sold too early but are eager to buy again.</li>
        </ol> The waiting areas for pullbacks are notoriously crowded!
        
If the market is in a trading range and you’re waiting for a breakout, you’ll have
to decide whether 
* to buy in anticipation of a breakout, 
* during a breakout, or 
* on a pullback after a valid breakout.

If you aren’t sure, consider entering in several steps:
buy a third of the planned position in anticipation, a third on a breakout, and a third
on a pullback.

>Whatever method you use, remember to apply the key risk management rule:
the distance from your entry to the protective stop, multiplied by position size can
never be more than 2 percent of your account equity (see Chapter 50).

Specific risk management tactics are different for trends and trading ranges.

<h3>22. Moving Averages <h5>p.74</h5></h3>



- the MA __slope__ identifies the direction of mass inertia. When it rises, it shows that the crowd is becoming more optimistic—bullish (and vice versa).  When prices rise above a moving average, the crowd is more bullish than before (and vice versa).

There are three main types of moving averages: 
- __Simple Moving Avergaes__ (SMA) : simple MAs have a fatal flaw—they change twice in response to each price.
- __Exponential Moving Averages__ (EMA) : gives greater weight to the latest data and responds to changes faster than a simple MA.

    An EMA has two major advantages over a simple MA.
       - First, it assigns greater weight to the last trading day.  The latest mood of the crowd is more important. 
       - Second, EMA does not drop old data the way a simple MA does. Old data slowly fades away, like a mood of the past lingering in a composite photo.
- __Weighted Moving Averages__

<h5>Choosing the Length of a Moving Average</h5>

- Most traders can use an EMA between 10 and 30 days. A moving average should not be shorter than 8 days to avoid defeating its purpose as a trend-following tool.
- Among the numbers I like are 22 because there are approximately 22 trading days in a month and 26—half of the number of trading weeks in a year.
- Don’t change indicator parameters while looking for trades. It is better to set your parameters and live with them.

<h5>Trading Rules</h5>

- <b>When EMA rises</b>, trade that market from the long side. <i>Buy when prices dip near the moving average</i>. Once you are long, place a protective stop below the latest minor low, and move it to the break-even point as soon as prices close higher.
- When the EMA falls, trade that market from the short side. 
- <b>When the EMA goes flat and only wiggles a little</b>, it identifies an aimless, trendless market. Do not trade using a trend-following method.

<h4>About Manual Testing</h4>

-  Open your spreadsheet, write down your system’s rules at the top of the page, and create columns for dates, prices, and signals.
- As you click forward, one day at a time, trends and trading ranges will slowly unfold and challenge you. At that point, you’ll be doing much more than testing a set of rules. Moving ahead one day at a time will test and improve your decision-making skills. This one-bar-at-a-time testing is vastly superior to what you can get from backtesting software.
- If one-bar-at-a-time testing shows positive results, start trading small positions with real money. 
-  Be sure to keep good records, and if your real-money results continue positive, start increasing the size of your trades. Do it in steps, all the way up to your normal trade size.

<h4>Paper Trading</h4>

- Paper trading means recording your buy and sell decisions and tracking them like real trades, but with no money at risk. 
- Beginners may start out paper trading, but most people turn to it after getting beat up by the markets.
- Psychology plays a huge role in how your trades turn out, and that’s where paper trading fails to deliver.
- There is only one good reason to paper trade—__to test your discipline as well as your system__.
>If you can download your data at the end of each day, do your homework, write
down your orders for the day ahead, watch the opening and record your entries,
then track your market each day, adjusting your profit targets and stops—if you can
do all of this for several months in a row, recording your actions, without skipping
a day—__then you have the discipline to trade real money__.

- An impulsive person who trades for entertainment will not be able to paper trade that way because it requires real work.

<h4>Three Key Demands for Every Trade</h4>



- __TRADE SETUP__ : write down the three key numbers for every trade: your __entry__,__target__, and __stop__. Before entering the market you need to decide how much you’ll pay, how much you’ll risk, and how much you expect to gain.
- __RISK MANAGEMENT__ : decide in advance how many dollars you’re prepared to risk on
this trade. Divide that amount by your risk per share—the distance from your entry to
your stop. __This will give you the number of shares you may trade__.
- __CLEAR TRADING STRATEGIES__ : Last but not least, every single trade must be based on a specific system or strategy. “It looks good to me” isn’t a system!
