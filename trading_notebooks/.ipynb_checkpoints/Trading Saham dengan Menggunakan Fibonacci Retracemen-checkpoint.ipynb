<h3>Bab 5, 6 : Support and Resistance, Tren</h3>

What to take from chapter 5 and 6 :

- __Aksioma__
>Manusia adalah penggerak harga. Karena manusia memiliki daya ingat, harga pun memiliki daya ingat. _Support dan resisten mereprentasikan ingatan dari harga_. Ini adalah aksioma dari suport dan resisten.
- __Bukti Observasi__
>Dengan mengamati pergerakan harga pada suatu saham dalam suatu rentang waktu tertentu, kita dapat mengambil kesimpulan bahwa harga bergerak naik dan turun didalam rentang interval yang terbatas, batas atas adalah resisten dan batas bawah adalah suport. Batas ini dapat ditembus tentunya. Tetapi dengan fakta bahwa harga dapat terkurung pada interval ini, trader dapat menentukan posisinya dan mengambil keuntungan dari pola pergerakan harga, yaitu level resisten menjadi tempat yang ideal untuk mengambil posisi jual, dan level suport menjadi tempat yang ideal untuk mengambil posisi beli.
- __Sifat Tidak Rigid Suport dan Resisten__ 
>Secara natural(?), resisten (support) yang tertembus dapat menjadi support (resisten) untuk pergerakan harga berikutnya.
- __Rules of Three__
>Syarat penembusan suport/resisten : Sebuah suport (resisten) dikatakan __tertembus__ apabila harga berhasil ditutup 3 hari di bawah (di atas) level tersebut __atau__ harga berhasil ditutup $3 \%$ di bawah (di atas) level tersebut.
- __Cara pemilihan level suport/resisten yang umum__ :
><ol>
    <li>Titik tertinggi/terendah harga-harga sebelumnya : dapat dipilih harga penutupan atau harga low (high).</li>  
    <li>Level-level psikologis : angka bulat atau angka2 yang umum dipilih manusia.</li>
    <li>Garis Tren : Dibuat dengan mengambungkan dua atau lebih titik terendah/tertinggi dari harga. Titik penarikan awal adalah titik terendah/tertinggi dari titik reversal.</li>
    <li>Fibonacci Retracement Level : Bab 7</li>
    <li>Gap, Moving Average, Candlestick, Pivot</li>
</ol>

- __Rule of Thumb Penentuan dengan Titik Tertinggi/Terendah__ :
><ol>
    <li>Gunakan titik ekstrim (low or high) bila memungkinkan.</li>  
    <li>Gunakan harga penutupan bila 1 terasa meragukan.</li>
    <li>SUport/resisten dikatakan kuat apabila level tersebut telah dikonfirmasi/telah diuji.</li>
    <li>Suport/resisten dapat dipilih sebagai daerah interval.</li>
    
</ol>


<h3>Bab 7 : Fibonacci Retracement</h3>

What to take from chapter 7 :

- __Fibonacci Retracement__ 
>alat analisis teknikal yang digunakan untuk mengukur tingkat pengembalian pergerakan harga bergerak dari suatu titik ekstrim atas/bawah ke titik ekstrim bawah/atas dari sebuah tren. Retracement diukur dalam presentase.

- __Presentase Level Retracement Fibonacci__ yang biasa digunakan sebagai suport dan resisten untuk pergerakan harga pada masa yang akan datang adalah :
><ol>
    <li>$23.6 \%$</li>  
    <li>$38.2 \%$</li>
    <li>$50.0 \%$</li>
    <li>$61.8 \%$</li>
    <li>$100 \%$</li>
    <li>$161.8 \%$</li>
    <li>$261.8 \%$</li>
    <li>$468.8 \%$</li>
</ol>

- __Cara membangun level-level Fibonacci Retracement__
><ol>
    <li>Menetukan Titik Puncak (Peak) dan Titik Dasar (Through) : kenali tren dan kapan tren berakhir</li>  
    <li>Gambar level-level Fibonacci Retracement</li>
</ol>

- __My Example__



- __Pergerakan harga dengan Panduan Fibonacci Retracement__

>Suport dan resisten adalah batas maya yang dibuat untuk memandu trader mengikuti arah pergerakan harga. Batas maya (suport dan resisten) yang ditarik/dihasilkan dnegan menggunakan Fibonacci retracement memiliki probabilitas (setidaknya menurut penulis) yang besar untuk menjadi suport/resisten yang kuat. Penggunaan Fibonacci retracement terbagi dalam dua kelompok : \n
>A. __Pergerakan harga antara dua titik__ : kasus-kasus yang sering terjadi
    <ol>
        <li>Ketika terjadi _trend reversal_, retracement $50 \%$ adalah target harga dengan probabilitas tertinggi</li>
        <li>Ketika saat _trend reversal_ harga langsung bergerak ke retracement $61.8 \%$, harga sangat berpotensi untuk langsung bergerak ke $38.2\%$</li>
        <li>Ketika terjadi _trend reversal_ tetapi terjadi konsolidasi di sekitar retracement $38.2\%$, maka retracement $61.8\%$ memiliki peluang untuk menjadi target pergerakan. Dalam kondisi ini retracement $50\%$ sering dilewati begitu saja.</li>
    </ol>
    

    

>B. __Pergerakan harga saat mencapai rekor harga baru__ : berikut level-level retracement yang menjadi target
    <ol>
        <li>Retracement $161.8 \%$ adalah target harga pertama</li>
        <li>Retracement $261.8 \%$, harga target berikutnya</li>
        <li>Retracement $423.6 \%$ adalah level kosolidasi saat harga bergerak terlalu kuat dan melewati $161.8 \%$ dan $261.8 \%$ tanpa kosolidasi yang berarti.</li>
    </ol>
